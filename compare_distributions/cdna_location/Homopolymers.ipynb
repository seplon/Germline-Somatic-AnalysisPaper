```python
def repeated_alphabet_lengths_threshold(s, threshold):
    if not s:
        return []

    lengths = []
    current_char = s[0]
    current_length = 1
    count=0

    for char in s[1:]:
        if char == current_char:
            current_length += 1
            count=count+1
        else:
            if current_length > threshold:
                lengths.append((current_char, current_length, count))
            current_char = char
            current_length = 1
            count=count+1

    if current_length > threshold:
        lengths.append((current_char, current_length))
    
    return lengths
```
